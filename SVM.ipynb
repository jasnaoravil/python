```python
#import the packages
import pandas as mypd
from sklearn import svm
#two things-svc for classification & svr for regression
#For parameter optimization
from sklearn.model_selection import GridSearchCV
```

```python
# Import the trainng data
mydata = mypd.read_csv("Iris_data.csv")
```

```python
mydata.head()
```

```python
# Seperate x's and y
x = mydata.iloc[:, 1:5]
y = mydata.Species
```

```python
# Specify the model
mymodel = svm.SVC()
```

```python
# Fit the model tot he data
mymodel = mymodel.fit(x,y)
```

```python
# Set up the parameter grid
myparamgrid = [{'kernel' : ['linear', 'poly', 'rbf'], 'degree' : [1, 2, 3], 'C': [ 0.001, 0.01, 0.1, 1, 5, 10, 100]}]
```

```python
# Set up the search
mysearch = GridSearchCV(mymodel, myparamgrid, scoring= 'accuracy', cv = 5, verbose= True)
```

```python
# Fit the search to the data
mysearch = mysearch.fit(x,y)
```

```python
# Best parameters ofthe model
mysearch.best_params_
```

```python
# Specify the optimum model
mymodel = svm.SVC(kernel= 'poly', degree= 2, C= 1, probability= True)
```

```python
# Fit the optimum model
mymodel = mymodel.fit(x, y)
```

```python
# Accuracy
accuracy = mymodel.score(x,y)
round(accuracy*100,2)
```

```python
# Predicted values
predy = mymodel.predict(x)
```

```python
# Actual versus Predicted values
mytable = mypd.crosstab(y, predy)
mytable
```

```python
# Import the test data
test = mypd.read_csv("Iris_test.csv")
```

```python
# Explore the test data
test.head()
```

```python
# Separate x's & y
testx = test.iloc[:, 1:5]
testy = test.Species
```

```python
# Accuracy on test data
test_accuracy = mymodel.score(testx, testy)
round(test_accuracy*100,2)
```

```python
# Predict the values for test data
predtest = mymodel.predict(testx)
```

```python
# Actual versus predicted for test data
mytesttable = mypd.crosstab(testy, predtest)
mytesttable
```

```python
# Make predicted vlaues as a dataframe
predtest = mypd.DataFrame(predtest, columns= ["predicted"])
```

```python
# predicted probabilities for test data
predprob = mymodel.predict_proba(testx)
```

```python
# make predicted probabiliteis as dataframe
predprob = mypd.DataFrame(predprob, columns= ["Setosa", "Versicolor", "Virginica"])
```
